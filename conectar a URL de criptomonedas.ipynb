Para ingestar la informacion de los historicos en Power BI (cambiando los formatos fecha), recurrimos a este codigo (uno por cada moneda):

```python
import  datetime
import requests
import pandas as pd
from datetime import datetime

base_currency = 'SOL'  # Esta es la sigla de la cripto a importar
quote_currency = 'USD' # Esta es la paridad de la cripto seleccionada antes
endpoint_url = 'https://ftx.com/api/markets'
request_url = f'{endpoint_url}/{base_currency}/{quote_currency}'

# Get all market data as JSON
all_markets = requests.get(endpoint_url).json()
# Convert JSON to Pandas DataFrame
df = pd.DataFrame(all_markets['result'])
df.set_index('name', inplace = True)
df. reset_index (drop = False , inplace = True )

# 1 day = 60 * 60 * 24 seconds
daily=str(60*60*24)
# Define start date as 2022-01-01
start_date = datetime(2022, 1, 1).timestamp()
# Get the historical market data as JSON
historical = requests.get(f'{request_url}/candles?resolution={daily}&start_time={start_date}').json()
# Convert JSON to Pandas DataFrame
df = pd.DataFrame(historical['result'])
# Convert time to date
df['date'] = pd.to_datetime(df['time']/1000, unit='s', origin='unix') 
# Remuevo columnas que no preciso para el analisis
df.drop(['startTime', 'time', 'low', 'high'], axis=1, inplace=True)
df
```
